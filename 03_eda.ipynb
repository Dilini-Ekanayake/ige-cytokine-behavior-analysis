```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # seaborn = prettier statistical plots, built on matplotlib

df = pd.read_csv('data/cleaned_data.csv')
log_cols = [c for c in df.columns if c.endswith('_log')]
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for ax, col in zip(axes.flatten(), log_cols + ['behavior_score']):
    ax.hist(df[col], bins=15, color='steelblue', edgecolor='white')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
for ax, col in zip(axes.flatten(), log_cols):
    ax.scatter(df[col], df['behavior_score'], alpha=0.6)
    ax.set_xlabel(col)
    ax.set_ylabel('behavior_score')
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df, x='behavior_group', y='behavior_score')
plt.title('Behavior score by group')
plt.show()
```
